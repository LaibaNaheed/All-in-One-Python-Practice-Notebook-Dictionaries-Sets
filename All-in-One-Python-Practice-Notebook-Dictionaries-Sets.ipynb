PART 1: Tuples:

```python
#Create a tuple
my_tuple = (10, 20, 30, 40, 50, 60, 70)
print("Original Tuple:", my_tuple)

# 1. Access tuple items using positive and negative indexes
print("Positive index [2]:", my_tuple[2])     # 30
print("Negative index [-2]:", my_tuple[-2])   # 60

# 2. Access tuple items using a range of positive and negative indexes
print("Range [1:4]:", my_tuple[1:4])          # (20, 30, 40)
print("Range [-4:-1]:", my_tuple[-4:-1])      # (40, 50, 60)

# 3. How to add items into a list (tuples are immutable)
my_list = list(my_tuple)
my_list.append(80)
print("After adding to list:", my_list)

# 4. How to change the tuple items
# (You must convert it to a list first, then back to a tuple)
my_list[0] = 5
my_tuple = tuple(my_list)
print("After modifying tuple:", my_tuple)

# 5. How to add tuple into a tuple
extra = (90, 100)
combined_tuple = my_tuple + extra
print("Tuple after adding another tuple:", combined_tuple)

# 6. Remove the tuple using delete and remove function
# You cannot remove items from a tuple, but you can delete the whole tuple
del combined_tuple
print("Tuple deleted successfully!")

# 7. Clear the tuple (create an empty tuple)
empty_tuple = ()
print("Cleared tuple:", empty_tuple)

# 8. Join two tuples using * and + operator
t1 = (1, 2, 3)
t2 = (4, 5, 6)
joined_plus = t1 + t2
repeated_tuple = t1 * 2

print("Joined using + :", joined_plus)
print("Repeated using * :", repeated_tuple)
```

2. Write a Python program that: Creates a dictionary to store student information (name,
age, and marks). Allows the user to add a new student entry. Updates the marks of an
existing student. Deletes a student from the dictionary. Prints all student details in a
readable format.

```python
students = {
    "Ali": {"age": 20, "marks": 85},
    "Sara": {"age": 21, "marks": 90}
}

print("\n8. Initial Student Dictionary:")
for name, info in students.items():
    print(name, "→", info)

# (a) Add a new student
students["Omar"] = {"age": 19, "marks": 88}
print("\nAfter adding a new student:", students)

# (b) Update marks of an existing student
students["Ali"]["marks"] = 95
print("\nAfter updating Ali's marks:", students)

# (c) Delete a student
del students["Sara"]
print("\nAfter deleting Sara:", students)

# (d) Print all student details nicely
print("\nAll Student Details:")
for name, info in students.items():
    print(f"Name: {name}, Age: {info['age']}, Marks: {info['marks']}")
```

PART 3: SETS IN PYTHON

```python
# Add items in sets
fruits = {"apple", "banana", "mango"}
fruits.add("orange")
print("1. After adding an item:", fruits)

# Add Any Iterable (like a list or tuple)
more_fruits = ["grape", "pineapple"]
fruits.update(more_fruits)
print("2. After adding iterable (list):", fruits)

#Add and join two sets
tropical = {"papaya", "coconut", "banana"}
joined_set = fruits.union(tropical)
print("3. Joined Set (Union):", joined_set)

#Remove items with different methods
numbers = {1, 2, 3, 4, 5}
print("\n4. Original set:", numbers)

numbers.remove(3)        # Removes specific element; gives error if not found
print("After remove(3):", numbers)

numbers.discard(10)      # No error even if element doesn’t exist
print("After discard(10):", numbers)

popped = numbers.pop()   # Removes random element
print("After pop():", numbers, "| Popped element:", popped)

numbers.clear()          # Removes all elements
print("After clear():", numbers)

# Deleting the entire set
del numbers
print("Set deleted successfully!\n")

#Join two sets using union() and update()
A = {1, 2, 3}
B = {3, 4, 5}

union_set = A.union(B)
print("5. Union using union():", union_set)

A.update(B)
print("After update() (A becomes):", A)

#Find the intersection of two sets
set1 = {1, 2, 3, 4}
set2 = {3, 4, 5, 6}
intersection = set1.intersection(set2)
print("\n6. Intersection:", intersection)

#Find the difference of two sets
difference = set1.difference(set2)
print("7. Difference (set1 - set2):", difference)

#Create a set of completed AI-related courses
completed_courses = {"Machine Learning", "Deep Learning", "NLP", "Computer Vision"}
print("\n8. Completed Courses:", completed_courses)

#Create another set of upcoming AI-related courses
upcoming_courses = {"Reinforcement Learning", "AI Ethics", "NLP", "Robotics"}
print("9. Upcoming Courses:", upcoming_courses)

#Find courses unique to completed_courses
unique_completed = completed_courses.difference(upcoming_courses)
print("10. Unique completed courses:", unique_completed)

#Set operations example
set_A = {1, 2, 3, 4, 5}
set_B = {4, 5, 6, 7}

print("\n11. Set Operations Example:")
print("Union:", set_A.union(set_B))
print("Intersection:", set_A.intersection(set_B))
print("Difference (A - B):", set_A.difference(set_B))

# Add and remove elements
set_A.add(8)
print("After adding 8 to A:", set_A)

set_A.remove(2)
print("After removing 2 from A:", set_A)

# Subset check
print("Is A a subset of B?", set_A.issubset(set_B))
```
